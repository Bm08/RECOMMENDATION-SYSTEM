```python
movie = movies.pivot(index='userId', columns='movie', values='rating').reset_index(drop=True)
```

```python
movie
```

```python
movie.index = movies.userId.unique()
```

```python
movie
```

```python
movie = movie.fillna(0)
```

```python
movie
```

```python
user_sim = 1 - pairwise_distances(movie.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df.index = movies.userId.unique()
user_sim_df.columns = movies.userId.unique()
```
